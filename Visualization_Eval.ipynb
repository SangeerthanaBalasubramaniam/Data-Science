```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/dayanandv/Data-Science/main/dataset/Boston_housing.csv")
df
```

```python
import matplotlib.pyplot as plt
% matplotlib inline
```

Part A - 1

```python
df['TAX'].plot(kind='hist', figsize=(10, 10), grid=False, edgecolor = 'black')
plt.xlabel("Tax", labelpad=10)
plt.ylabel("Count", labelpad=10)
plt.show()
```

Part A - 2

```python
import seaborn as sns
```

```python
sns.jointplot(x="PRICE", y="RM", data=df, kind="reg", color="#0079b5")
plt.show()
```

Part A - 2

```python
sns.heatmap(df[['B', 'TAX', 'PRICE']], vmin=0, vmax=700)
plt.show()
```

```python
df1 = pd.read_excel("https://github.com/dayanandv/Data-Science/raw/main/dataset/unemployment.xlsx")
df1
```

Part B - 1

```python
plt.figure(figsize=(16,5))
p = ['#b00054', '#b35000']
sns.stripplot(x = df1.Age, palette = p, y = df1.Unemployed , jitter = 0.3, hue = df1.Gender)
plt.show()
```

Part B - 2

```python
df2 = pd.read_csv("https://raw.githubusercontent.com/dayanandv/Data-Science/main/dataset/Boston_housing.csv")
ax = sns.violinplot(x = df2["AGE"])
```

Part B - 3

```python
plt.figure(figsize = (15, 10))
sns.boxplot(x = df1['Age'], y=df1['Unemployed'], palette="Set1")
plt.show()
```

Part B - 4

```python
data = df1.groupby("Gender").size()
data.plot(kind='pie', subplots=True, figsize=(8, 8),explode=[0.1,0.1],startangle=90,label=" ",autopct='%1.1f%%',shadow=True)
plt.show()
```
